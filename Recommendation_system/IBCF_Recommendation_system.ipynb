# Model Building and Identifying the similar users for recommedation

```python
item_item_recommendation.values
```

```python
item_item_recommendation = pairwise_distances(X = item_item_recommendation.values)
item_item_recommendation
```

```python
item_item_df = pd.DataFrame(item_item_recommendation)
item_item_df.head()
```

```python
item_item_df.shape
```

```python
# Replacing the index
item_item_df.index = book_data['Book_title'].unique()
item_item_df.head()
```

```python
# Replace the columns
item_item_df.columns = book_data['Book_title'].unique()
item_item_df.head()
```

```python
item_item_similarity = 1 - pairwise_distances(X = item_item_df.values, metric='cosine')
item_item_similarity
```

```python
item_item_df = pd.DataFrame(item_item_similarity)
item_item_df.head()
```
