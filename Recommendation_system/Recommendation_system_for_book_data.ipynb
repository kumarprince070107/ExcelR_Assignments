# Model Building and Identifying the similar users for recommedation

```python
user_user_recommendation.values
```

```python
user_user_similarity = pairwise_distances(X = user_user_recommendation.values)
user_user_similarity
```

```python
user_user_df = pd.DataFrame(user_user_similarity)
user_user_df.head()
```

```python
user_user_df.shape
```

```python
# Replacing thwe index
user_user_df.index = book_data['User_ID'].unique()
user_user_df.head()
```

```python
# Replace the columns
user_user_df.columns = book_data['User_ID'].unique()
user_user_df.head()
```

```python
user_user_similarity = 1 - pairwise_distances(X = user_user_recommendation.values, metric='cosine')
user_user_similarity
```

```python
user_user_df = pd.DataFrame(user_user_similarity)
user_user_df.head()
```
