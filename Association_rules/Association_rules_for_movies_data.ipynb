# Import Libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from sklearn.preprocessing import LabelEncoder
```

# Import Dataset

```python
movies_data=pd.read_csv('my_movies.csv')
movies_data.head()
```

# Data Understanding

```python
movies_data.shape
```

```python
movies_data.isnull().sum()
```

```python
movies_data.info()
```

# Data Preprocessing

```python
movie_data=movies_data.drop(columns=['V4','V5'],axis=1)
movie_data.head()
```

```python
movie_data=movie_data.dropna()
```

```python
 movie_data.isnull().sum()
```

```python
le=LabelEncoder()
```

```python
movie_data['V1']=le.fit_transform(movie_data.V1)
movie_data['V2']=le.fit_transform(movie_data.V2)
movie_data['V3']=le.fit_transform(movie_data.V3)
```

```python
movie_data.head()
```

```python
def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
movie_data[['V1','V2','V3']]=movie_data[['V1','V2','V3']].applymap(encode)
movie_data.head()
```

# Model Building

```python
# With min_support=0.10 and metric=confidence,threshold=0.10
```

```python
frequent_items=apriori(df=movie_data,min_support=0.10,use_colnames=False)
```

```python
rules=association_rules(df=frequent_items,metric='confidence',min_threshold=0.10)
rules
```

```python
rules.sort_values(by='support',axis=0,ascending=False)
```

```python
rules[rules.lift>1]
```

```python
# With min_support=0.15 and metric=confidence,threshold=0.7
```
