# Import Libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

# Import Dataset

```python
book_data=pd.read_csv('book.csv')
book_data.head()
```

# Data Understanding

```python
book_data.shape
```

```python
book_data.isnull().sum()
```

# Model Building

```python
# With min_support=0.10 and metric=confidence,threshold=0.10
```

```python
frequent_items=apriori(df=book_data,min_support=0.10,use_colnames=False)
```

```python
rules=association_rules(df=frequent_items,metric='confidence',min_threshold=0.10)
rules
```

```python
rules.sort_values(by='support',axis=0,ascending=False)
```

```python
rules[rules.lift>1]
```
